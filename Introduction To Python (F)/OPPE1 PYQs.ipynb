```
# Python OPPE 1 Problems Compilation
# This file contains a collection of Python programming problems and their solutions
# extracted from the provided PDF exam sets. Each problem is separated by a
# comment block for clarity and can be run independently.

from collections import Counter
import math

# ----------------------------------------------------------------------
# May Term 2024 - Set 1
# ----------------------------------------------------------------------

# Q1.1: Find the absolute difference between the sum of two integers and the sum of their squares.
def abs_diff_between_sum_and_sum_of_squares(a: int, b: int) -> int:
    """Calculates the absolute difference between the sum of two integers and the sum of their squares."""
    sum_val = a + b
    sum_sq_val = a**2 + b**2
    return abs(sum_val - sum_sq_val)

# Q1.2: Swap parts of an odd-length string before and after the middle three characters.
def swap_except_middle_three(s: str) -> str:
    """Swaps the parts of an odd-length string before and after the middle three characters."""
    mid_index = len(s) // 2
    middle_three = s[mid_index-1 : mid_index+2]
    before_middle = s[:mid_index-1]
    after_middle = s[mid_index+2:]
    return after_middle + middle_three + before_middle

# Q1.3: Interleave three lists of the same length.
def interleave_lists(list1, list2, list3):
    """Interleaves three lists of the same length."""
    interleaved = []
    for i in range(len(list1)):
        interleaved.extend([list1[i], list2[i], list3[i]])
    return interleaved

# Q1.4: Determine if a given integer has more than 5 unique digits.
def has_more_than_5_unique_digits(num: int) -> bool:
    """Determines if a given integer has more than 5 unique digits."""
    return len(set(str(abs(num)))) > 5

# Q1.5: Find the final position of a point given initial position, velocity, and time.
def final_position(pos: tuple, vel: tuple, time: int) -> tuple:
    """Calculates the final position of a point after a given time."""
    final_x = pos[0] + vel[0] * time
    final_y = pos[1] + vel[1] * time
    return (final_x, final_y)

# Q1.6: Remove keys from a dictionary that are not present in a given list (in-place).
def remove_keys_not_in_list(d: dict, l: list) -> None:
    """Removes keys from a dictionary that are not in a given list, in-place."""
    keys_to_remove = [key for key in d if key not in l]
    for key in keys_to_remove:
        del d[key]

# ----------------------------------------------------------------------
# Python OPPE 1 - Set 2
# ----------------------------------------------------------------------

# Q2.1: Square an integer and clip the result to a threshold.
def square_and_clip(x: int, threshold: int) -> int:
    """Squares an integer and clips the result if it exceeds the threshold."""
    squared = x ** 2
    return min(squared, threshold)

# Q2.2: Convert first half of a string to lowercase and second half to uppercase.
def lowercase_first_half_and_uppercase_second_half(s: str) -> str:
    """Converts the first half of a string to lowercase and the second half to uppercase."""
    mid = len(s) // 2
    return s[:mid].lower() + s[mid:].upper()

# Q2.3: Add the middle element of an odd-length list to both ends (in-place).
def add_the_middle_element_to_both_ends(l: list) -> None:
    """Adds the middle element of an odd-length list to both its ends, in-place."""
    mid_index = len(l) // 2
    middle_element = l[mid_index]
    l.insert(0, middle_element)
    l.append(middle_element)

# Q2.4: Return the number of unique digits common in two integers.
def number_of_unique_common_digits(n1: int, n2: int) -> int:
    """Returns the number of unique digits common in two integers."""
    return len(set(str(abs(n1))) & set(str(abs(n2))))

# Q2.5: Calculate the Manhattan distance from point a to c via point b.
def manhattan_distance_via_b(a: tuple, b: tuple, c: tuple) -> int:
    """Calculates the Manhattan distance from point a to c via point b."""
    dist_ab = abs(a[0] - b[0]) + abs(a[1] - b[1])
    dist_bc = abs(b[0] - c[0]) + abs(b[1] - c[1])
    return dist_ab + dist_bc

# Q2.6: Create a dictionary with indices as keys and names from a list as values.
def create_indexed_dict(names: list) -> dict:
    """Creates a dictionary with indices as keys and names from a list as values."""
    return {i: name for i, name in enumerate(names)}

# ----------------------------------------------------------------------
# Python OPPE 1 - Set 3
# ----------------------------------------------------------------------

# Q3.1: Check if sides form a right triangle with even perpendicular sides.
def is_right_triangle_with_even_sides(a: int, b: int, c: int) -> bool:
    """Checks if a, b, c form a right triangle with even perpendicular sides."""
    return a**2 + b**2 == c**2 and a % 2 == 0 and b % 2 == 0

# Q3.2: Check if odd indices are letters and even indices are digits.
def is_odd_indices_alpha_and_even_indices_digits(string: str) -> bool:
    """Checks if odd-indexed chars are letters and even-indexed chars are digits."""
    return string[::2].isdigit() and string[1::2].isalpha()

# Q3.3: Swap values at even and odd indices in a list (in-place).
def swap_even_and_odd_indices(l: list) -> None:
    """Swaps the values at even and odd indices in a list, in-place."""
    for i in range(0, len(l) - 1, 2):
        l[i], l[i+1] = l[i+1], l[i]

# Q3.4: Find unique characters present in the first string but not the second.
def unique_chars_present_in_first_not_in_second(s1: str, s2: str) -> set:
    """Returns a set of unique characters present in s1 but not in s2."""
    return set(s1) - set(s2)

# Q3.5: Create a tuple with 'a' repeated 'b' times and 'b' repeated 'a' times.
def repeat(t: tuple) -> tuple:
    """Creates a new tuple by repeating elements of the input tuple."""
    a, b = t
    return tuple([a] * b + [b] * a)

# Q3.6: Create a dictionary with numbers from 1 to n as keys and their squares as values.
def num_squares(n: int) -> dict:
    """Creates a dictionary with numbers 1 to n as keys and their squares as values."""
    return {i: i**2 for i in range(1, n + 1)}

# ----------------------------------------------------------------------
# Sep Oppe 1 - Set 1
# ----------------------------------------------------------------------

# Q4.1: Return the sum of squares and absolute difference of squares of two numbers.
def sum_squares_abs_diff_squares(a: int, b: int) -> tuple:
    """Returns a tuple with the sum of squares and the absolute difference of squares."""
    sum_sq = a**2 + b**2
    abs_diff_sq = abs(a**2 - b**2)
    return (sum_sq, abs_diff_sq)

# Q4.2: Check if a string is a palindrome of odd length.
def is_odd_length_palindrome(s: str) -> bool:
    """Checks if a string is a palindrome with odd length."""
    return s == s[::-1] and len(s) % 2 != 0

# Q4.3: Remove elements at two specified indices from a list (in-place).
def remove_elements_at_two_indices(l: list, i1: int, i2: int):
    """Removes elements at two specified indices from a list, in-place."""
    # Delete the element with the larger index first to avoid shifting issues.
    if i1 > i2:
        del l[i1]
        del l[i2]
    else:
        del l[i2]
        del l[i1]

# ----------------------------------------------------------------------
# Sep Oppe 1 - Set 2
# ----------------------------------------------------------------------

# Q5.1: Check if a number is a positive odd or a negative even.
def is_positive_odd_or_negative_even(n: int) -> bool:
    """Checks if a number is a positive odd or a negative even."""
    is_pos_odd = n > 0 and n % 2 != 0
    is_neg_even = n < 0 and n % 2 == 0
    return is_pos_odd or is_neg_even

# Q5.2: Check if a string is enclosed with double quotes and has double quotes inside.
def within_and_has_double_quotes(s: str) -> bool:
    """Checks if a string starts/ends with double quotes and has them inside."""
    return s.startswith('"') and s.endswith('"') and '"' in s[1:-1]

# Q5.3: Replace the middle element of a tuple with 'n' copies of itself.
def replace_middle_with_n_times_middle(t: tuple, n: int) -> tuple:
    """Replaces the middle element of an odd-length tuple with n copies of itself."""
    mid_index = len(t) // 2
    middle_element = t[mid_index]
    return t[:mid_index] + (middle_element,) * n + t[mid_index+1:]

# ----------------------------------------------------------------------
# Sep Oppe 1 - Set 3
# ----------------------------------------------------------------------

# Q6.1: Check if a number is divisible by exactly one of two other numbers.
def div_by_exactly_one(num: int, a: int, b: int) -> bool:
    """Checks if num is divisible by exactly one of a or b."""
    div_by_a = num % a == 0
    div_by_b = num % b == 0
    return div_by_a != div_by_b  # XOR logic

# Q6.2: Convert a date string from mm-dd-yy to yy-dd-mm.
def mm_dd_yy_to_yy_dd_mm(date: str) -> str:
    """Converts a date string from mm-dd-yy to yy-dd-mm format."""
    parts = date.split('-')
    return f"{parts[2]}-{parts[1]}-{parts[0]}"

# Q6.3: Increment a dictionary value for a key, but not beyond a specified limit.
def increment_value_with_max_limit(d: dict, key: str, inc: int, limit: int):
    """Increments a dictionary value for a key, capped at a limit."""
    d[key] = min(d[key] + inc, limit)
```
